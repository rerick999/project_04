```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create an array that contains the number of users each language has
accuracy = [94, 100, 100, 100, 100]
models = ["Logistic Regression", "Decision Tree", "Random Forest", "Gradient Boost Classifier", "XGBoost"]
x_axis = np.arange(len(accuracy))
```

```python
# Create a bar chart based upon the above data
tick_locations = [value for value in x_axis]

plt.bar(x_axis, accuracy, color=['red', 'orange', 'blue', 'green', 'purple'], alpha=0.5, align="center")
plt.title("Model Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.xticks(tick_locations, models, rotation = 90)
plt.tight_layout()
plt.savefig("Model_Comparison.png")
plt.show()
```
